```python
import pandas as pd
import matplotlib.pyplot as plt
import datetime
```

```python
df = pd.read_csv('/Users/pankeshpatel/Desktop/bmw-material-planner-assistant/backend/total_qty.csv') 
dataframe = pd.DataFrame(df, columns=['material', 'demand_date', 'max', 'min', 'mean', 'safety stock'])
dataframe
```

```python
font = {
        'color':  'black',
        'weight': '10',
        'size': 20,
        }


# Create figure
fig = plt.figure(figsize=(10, 6))

# Plot time series
plt.plot(dataframe['demand_date'], dataframe['max'], marker='o', label='Max of Total Quatity', linewidth=4, alpha=0.7)
plt.plot(dataframe['demand_date'], dataframe['min'], marker='o', label='Min of Total Quatity', linewidth=4, alpha=0.7)
plt.plot(dataframe['demand_date'], dataframe['mean'], marker='o', label='Average of Total Quatity', linewidth=4, alpha=0.7)
plt.plot(dataframe['demand_date'], dataframe['safety stock'], marker='o', label='Safety Stock', linewidth=4, alpha=0.7)


plt.title(label = 'HealthScore Analysis of Material {} of Next 10 days'.format(dataframe['material'][0]),
         fontsize=25,
         color="black",
         pad='2.0',
         fontweight=10,
         loc="center")

plt.xlabel('Date', fontdict=font)
plt.ylabel('Quantity of Material', fontdict=font)

# # Add legend
plt.legend(loc='best')


# # Auto space
plt.tight_layout()

# # Display plot
plt.show() 
```
